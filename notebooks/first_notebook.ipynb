```python
import string
import numpy as np
import re
from nltk.corpus import stopwords
from collections import Counter
class naive_bayes_classifier():

    def __init__(self, min_freq=1):
         self.min_freq = min_freq
     
    def preprocess_text(self, text):
        text = text.lower()
        translator = str.maketrans('', '', string.punctuation)
        clean_text = text.translate(translator)
        tokens = re.findall(r'\w+', clean_text)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [token for token in tokens if token not in stop_words]
        return filtered_tokens

    def vocabulary(self, text):
        vocab = {}
        index = 0
        vocab["<UNK>"] = index
        tokens = [] 
        for sentence in text:
            y = self.preprocess_text(sentence)
            tokens.extend(y)
        freq = Counter(tokens)
        for items, counts in freq.items():
            if counts >= self.min_freq:
                index += 1
                vocab[items] = index
        return vocab

    def calculate_class_probabilities(self, labels):
        n = len(labels)
        class_count = {}
        for label in labels:
            if label not in class_count:
                class_count[label] = 0
            class_count[label] += 1
        self.class_priors = {}
        for clas in class_count:
            self.class_priors[clas] = class_count[clas]/n
        return self.class_priors
    
    def calculate_word_probabilities(self, texts, labels, vocab):
        vocab_len = len(vocab)
        classes = set(labels)
        count = {}
        total_word = {}
        prob_cal = {}
        self.word_likelihoods = {}
        for clas in classes:
            total_word[clas] = 0
            count[clas] = np.zeros(vocab_len)
        for text, label in zip(texts, labels):
            tokens = self.preprocess_text(text)
            for token in tokens:
                idx = vocab[token] if token in vocab else vocab["<UNK>"]
                count[label][idx] += 1
                total_word[label] += 1
        for clas in classes:
            prob_cal[clas] = (count[clas] + 1)/(total_word[clas] + vocab_len) 
            temp = {}      
            for word in vocab:
                index = vocab[word]
                prob = prob_cal[clas][index]
                temp[word] = prob
            self.word_likelihoods[clas] = temp
        return self.word_likelihoods

    def fit(self, texts, labels):
        self.vocab = self.vocabulary(texts)
        self.class_priors = self.calculate_class_probabilities(labels)
        self.word_likelihoods = self.calculate_word_probabilities(texts, labels, self.vocab)


    def predict(self, sentence):
        tokens = self.preprocess_text(sentence)
        scores = {}
        for clas in self.class_priors:
            score = np.log(self.class_priors[clas])
            for token in tokens:
                if token not in self.vocab:
                    token = "<UNK>"
                score += np.log(self.word_likelihoods[clas][token])
            scores[clas] = score
        class_with_max_score = max(scores, key=scores.get)
        return class_with_max_score


def main():
    text = ["i love coffee!", "I love tea.", "Do I love any of these? ha ha ha, nice try, but No!", "Sheldon Lee cooper has an iq of 187 but doesn't like coffee.", "he likes tea"]
    labels = [1, 1, 0, 0, 1]
    classifier = naive_bayes_classifier()
    classifier.fit(text, labels)
    print(classifier.predict("kirtiamy"))

if __name__ == "__main__":
    main()
```

```python
import numpy as np
from collections import Counter
import torch
import torch.nn as nn
import torch.optim as optim
class tfidf_vectorizer():
    def __init__(self, min_freq=2):
        self.min_freq = min_freq
        self.vocab = None
        self.new_list = []
        self.df = None
        self.idf = None
        self.n_docs = None
    
    def vocabulary(self):
        self.texts = [
        "I like coffee",
        "I like tea",
        "I like coffee and tea"
        ]
        tokens = []
        [tokens.extend(sentence.lower().split()) for sentence in self.texts]
        freq = Counter(tokens)
        self.n_docs = len(self.texts)
        # creating vocabulary
        index = 0
        self.vocab = {}
        self.vocab["<UNK>"] = index
        for items, counts in freq.items():
            if counts >= self.min_freq:
                index += 1
                self.vocab[items] = index
    
    def new_sent(self, new_sentence):
        word = new_sentence.lower().split()
        new_list = []
        for w in word:
            if w in self.vocab:
                new_list.append(self.vocab[w])
            else:
                new_list.append(self.vocab["<UNK>"])
        return new_list
    
    def bag_of_words(self, sentence):
        nw = self.new_sent(sentence)
        bow = np.zeros(len(self.vocab))
        for indices in nw:
                bow[indices] += 1
        return bow

    def df_(self):
        self.df = np.zeros(len(self.vocab))
        for sentence in self.texts:
            bow = self.bag_of_words(sentence)
            for i in range(len(bow)):
                if bow[i] > 0:
                    self.df[i] += 1

    def idf_(self):
        self.idf = np.zeros(len(self.vocab))
        for i in range(len(self.df)):
            df__ = self.df[i]
            self.idf[i] = np.log((self.n_docs+1)/(df__+1)) + 1

    def tf_idf(self, sentence):
        bow = self.bag_of_words(sentence)
        tf_idf_ = bow * self.idf
        return tf_idf_
    
    def cosine_similarity(self, sentence1, sentence2):
        x = self.tf_idf(sentence1)
        y = self.tf_idf(sentence2)
        if np.linalg.norm(x) == 0 or np.linalg.norm(y) == 0: 
            return 0
        else:
            cos_sim = np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))
            return cos_sim

class text_classifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.linear = nn.Linear(input_dim, num_classes)
    
    def forward(self, x):
        return self.linear(x)
    
def train(model, vectorizer, texts, labels, epochs):
    model.train()
    loss_fn = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(),lr=0.01)
    for epoch in range(epochs):
        total_loss = 0
        for text, label in zip(texts, labels):
            vector = vectorizer.tf_idf(text)
            vector = vector.reshape(1, -1)
            tensor = torch.tensor(vector, dtype=torch.float32)
            output = model(tensor)
            
            label = torch.tensor(label, dtype=torch.long)
            label = label.reshape(1, -1)
            loss = loss_fn(output, label)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
        average_loss = total_loss / len(texts)
        print(epoch, average_loss)

def main():
    vectorizer = tfidf_vectorizer()
    vectorizer.vocabulary()
    vectorizer.df_()
    vectorizer.idf_()
    tf_idf_ = vectorizer.tf_idf("I like coffee")
    tf_idf_ = tf_idf_.reshape(1, -1)
    sim = vectorizer.cosine_similarity("I LOVE coffee", "I love milk")
    print(sim)
    model = text_classifier(5, 2)           # nn.Linear(vocab  size, number of classes)
    texts = [
    "I love coffee",
    "I hate coffee",
    "coffee is great",
    "coffee is terrible",
    "I like tea",
    "I dislike tea"
        ]
    labels = [
    1,  # I love coffee
    0,  # I hate coffee
    1,  # coffee is great
    0,  # coffee is terrible
    1,  # I like tea
    0   # I dislike tea
                ] 
    train(model, vectorizer, texts, labels, epochs=5)    

if __name__ == "__main__":
    main()
```

```python
def is_prime(n, divisor=2):
    if n < 2:
        return False
    
    if n == 2:
        return True
    
    if divisor > math.sqrt(n):
        return True
    
    if n%divisor == 0:
        return False
    
    return is_prime(n, divisor+1)

print(is_prime(27))
```

```python
def f(x):
    return 2*x
a = int(input("enter lower limit of integral"))
b = int(input("enter upper limit of integral"))
n = int(input("enter number of steps"))
h = (b-a)/n if n>0 else print("Invalid n, n must be greater than 0")
s = 0
for i in range(1, n):
    s += f(a + i*h)
integral = (h/2)*(f(a) + 2*s + f(b))
print(f"final integral is {integral:.4f}")
```

```python
s = input("enter a string")
s = s.lower().strip()
final = []
for i in range(97, 123):
    final.append([chr(i), 0])
for char in s:
    index = ord(char)
    if 97 <= index <= 122:
        final[index-97][1] += 1
for x,y in final:
    if y>0:
        print(f"{x}:{y}")
```

```python
num_list = [1,2,4,5,6,8,7,8,9,0,4,5,5,6,4,3,2,6,7,8,0,1,2,6,3,4,7]
seen = set()
temp = []
for i in num_list:
    if i not in seen:
        temp.append(i)
        seen.add(i)
temp.sort(key = lambda x : x%2==0)     
print(temp)  
```
